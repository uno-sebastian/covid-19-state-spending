```python
"""
The two most useful dataframes are df3 and df4.
df3 has all data less irrelevant columns. This data needs another cleaning step to remove nans from desired columns.
df4 has all of the most stable and relevant data for deaths and has been cleared of nans
Final output is the df4 as Most_Stable_Data.csv
"""
```

```python
import pandas as pd
file = pd.read_csv('covidtracking_dataframe.csv')
```

```python
df = pd.DataFrame(file)
list(df.columns)
df.head()
```

```python
df1 = df.drop(columns=['grade','score','totalTestResultsSource','positiveScore','negativeScore','negativeRegularScore','commercialScore','hash','dataQualityGrade','lastUpdateEt','checkTimeEt','dateChecked','dateModified','fips','probableCases'])
df1.head()
```

```python
nan = {}
for column in df1.columns:
    num_nan = df1[column].isna().sum()
    print(f'{column} # of NAN: {num_nan}')
    nan[column] = num_nan
```

```python
maxi = 0
for param in nan:
    if nan[param] > maxi:
        maxi = nan[param]
        print(f'Current Max: {nan[param]}')
```

```python
df2 = df1.drop(columns=['positiveTestsPeopleAntigen','totalTestsPeopleAntigen','totalTestsAntigen','pending'])
df2.sort_values(by=['date'])
df2
```

```python
df3 = df2.dropna(subset=['death'])
df3
```

```python
df4 = df3[['date','state','positive','negative','death','totalTestResults','hospitalized']]
df4.dropna(inplace=True)
df4.reset_index()
df4.to_csv('Most_Stable_Data.csv')
```
